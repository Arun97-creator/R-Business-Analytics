```r
# Above Order is not Ideal
# Forcats Packages
install.packages('forcats')
library(forcats)
```

```r
# Convert to factor  and order by frequency----
#?fct_infreq # Help
df$Category3 <- forcats::fct_infreq(df$Category) # New Column3
str(df$Category3)
summary(df$Category3)
plot(df$Category3)
```

```r
#Lump Function ?fct_lump
df$Category4 <- fct_lump(df$Category, n=3) # I just want to see the top 3 levels
summary(df$Category4)
plot(df$Category4)
```

```r
# Descending Order
df$Category4 <- fct_infreq(fct_lump(df$Category, n=3)) # I just want to see the top 3 levels
summary(df$Category4)
plot(df$Category4)
```

```r
# You can relevl by hand
# ?fct_relevel

df$Category5 <- fct_relevel(df$Category4, 'Salmon and Wheat Bran Salad', 'general','Chicken','Other')
plot(df$Category5)
```
